```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
#Dropping index number 122
df=df.dropna().reset_index(drop=True)
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df['Region']=df['Region'].astype(int)
```

```python
df.info()
```

```python
df.columns
```

```python
#Removing blank spaces in the names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

### Changing required datatypes to int and float type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# Again changing required datasets into float using different method
Objects=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
Objects
```

```python
#only classes has null, thus every other Object can be turned to float
for i in Objects:
    if i != 'Classes' :
        df[i]=df[i].astype(float) 
```

```python
# Monthly Fire Analysis of Sidi- Bel Regions
dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```

```python
# Monthly Fire Analysis of Bejaia
dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Bejaia",weight='bold')
```
